## TASK 4: Image to Pencil Sketch with Python

#### We need to read the image in RBG format and then convert it to a grayscale image.

###### Steps used in the notebook
        
        1. Importing libraries
        2. Read image in RGB format
        3. Convert to gray image
        4. Invert the image then blur and then invertedblur
        5. Convert the image into Pencil Sketch

### 1. Importing libraries

```python
import cv2
```

```python
from PIL import Image
```

### 2. Read image in RGB format

```python
image = cv2.imread('dog.jfif')
```

```python
display(Image.fromarray(image))
```

### 3. Convert to gray image

```python
gry_filter = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
display(Image.fromarray(gry_filter))
```

### 4. Invert the image then blur and then invertedblur

```python
inverted = 255-gry_filter
```

```python
blurred = cv2.GaussianBlur(inverted,(21,21),0)
```

```python
invertedblur = 255-blurred
```

### 5. Convert the image into Pencil Sketch

```python
pencilsketch = cv2.divide(gry_filter, invertedblur,scale=256.0)
```

```python
cv2.imwrite('dog.jpg',pencilsketch)
```
